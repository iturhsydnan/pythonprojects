**Perform unigram analysis and removal of stopwords:**

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
```

```python
stop
```

```python
stop.extend(['rated', 'n', 'nan', 'x', 'RATED', 'Rated'])
```

```python
stop
```

```python
rev3= reviews_tokens[3]
rev3
```

```python
print([token for token in rev3 if token not in stop])
```

```python
reviews_tokens_clean= reviews_tokens.apply(lambda each_review: [token for token in each_review if token not in stop])
```

```python
reviews_tokens_clean
```

```python
type(reviews_tokens_clean)
```

```python
total_reviews2= list(reviews_tokens_clean)
```

```python
tr1=[]
for review in total_reviews2:
    for word in review:
        tr1.append(word)
```

```python
tr1
```

```python
from nltk import FreqDist
```

```python
fd= FreqDist()
```

```python
for word in tr1:
    fd[word]=fd[word]+1
```

```python
fd.most_common(20)
```

```python
fd.plot(20)
```

**Performing bigram and tigram analysis:**

```python
from nltk import FreqDist, bigrams, trigrams
```

```python
bi_grams= bigrams(tr1)
```

```python
fd_bigrams= FreqDist()
for bigram in bi_grams:
    fd_bigrams[bigram]= fd_bigrams[bigram]+1
```

```python
fd_bigrams.most_common(20)
```

```python
fd_bigrams.plot(20)
```

```python
tri_grams=trigrams(tr1)
```

```python
fd_trigrams= FreqDist()
for trigram in tri_grams:
    fd_trigrams[trigram]= fd_trigrams[trigram]+1
```

```python
fd_trigrams.most_common(20)
```

```python
fd_trigrams.plot(20)
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator= Nominatim(user_agent= 'app', timeout=None)
```

```python
lat=[]
long=[]
for location in locations['Name']:
    location = geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        long.append(np.nan)
    else:
        lat.append(location.latitude)
        long.append(location.longitude)
```

```python
locations['latitude'] = lat
locations['longitude']= long
```

```python
def get_heatmap(cuisine):
    cuisine_df=df[df['cuisines'].str.contains(cuisine)]
    cuisine_rest_count= cuisine_df['location'].value_counts().reset_index().rename(columns={'index': 'count', "location": "Name"})
    heatmap_df=cuisine_rest_count.merge(locations, on='Name', how='left')
    print(heatmap_df.head(4))
    
    basemap=folium.Map()
    HeatMap(heatmap_df[['latitude', 'longitude', "count"]]).add_to(basemap)
    return basemap
```

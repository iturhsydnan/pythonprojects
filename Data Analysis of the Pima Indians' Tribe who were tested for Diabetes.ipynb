```python
corr_matrix= pima.iloc[:, 0:8].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Heatmap is basically colour coded matrix, showing correlation between the variables. Correlation shows the realtionship between two or more variables. The darkest colour means negative correlation and lightest color or white means 1 or positive. The diagonal line of '1' shows positive relation. 
The ones with negative correlation are SkinThickness and Pregnancies(-0.082), SkinThickness and Age (-0.11), Insulin and Pregnancies (-0.074), Insulin and Age (-0.042), DiabetesPedigreeFunction and Pregnancies (-0.034). The negative correlation shows that one variable has nothing to do with the other. For example, for this heatmap, Age is not an important factor for SkinThickness.  
Rest of the correlation between two different variable shows weak positive correlation. 
The correlation of Age and Pregnancies is 0.54, it is inside a bright orange-red box. It means that Age particulary do not have a strong relationship with Pregnancies, but still Age does work as a factor in Pregnancies.

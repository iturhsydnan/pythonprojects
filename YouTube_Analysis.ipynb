**Sentiment Analysis:**

```python
from textblob import TextBlob
```

```python
comments.head(6)
```

```python
TextBlob("Say hi to Kong and maverick for me").sentiment
```

**Polarity:** This is a floating-point number that represents the sentiment polarity of the text. It can range from -1 (very negative) to 1 (very positive). A polarity of 0 typically indicates a neutral sentiment.

**Subjectivity:** This is another floating-point number that represents the subjectivity of the text. It can range from 0 to 1, where 0 is very objective (factual) and 1 is very subjective (opinionated).

```python
TextBlob("Say hi to Kong and maverick for me").sentiment.polarity
```

```python
polarity = []
for comment in comments['comment_text']:
  try:
    polarity.append(TextBlob(comment).sentiment.polarity)
  except:
    polarity.append(0)
```

```python
len(polarity)
```

```python
comments['polarity'] = polarity
```

```python
comments.head(15)
```

**Wordcloud Analysis:** Graphical representation of text frequency

```python
filter1= comments['polarity']==1
pos_comment = comments[filter1]
pos_comment
```

```python
filter2= comments['polarity']== -1
neg_comment= comments[filter2]
neg_comment
```

```python
from wordcloud import WordCloud, STOPWORDS
```

STOPWORDS are those words that do not make any sense in the analysis aka meaningless words

```python
set(STOPWORDS)
```

```python
comments['comment_text']
```

```python
type(comments['comment_text'])
```

```python
total_pos_comment= ' '.join(pos_comment['comment_text'])
```

```python
wordcloud= WordCloud(stopwords=set(STOPWORDS)).generate(total_pos_comment)
```

```python
plt.imshow(wordcloud)
plt.axis('off')
```

```python
total_neg_comment= ' '.join(neg_comment['comment_text'])
```

```python
wordcloud1= WordCloud(stopwords=set(STOPWORDS)).generate(total_neg_comment)
```

```python
plt.imshow(wordcloud1)
plt.axis('off')
```

**EMOJI ANALYSIS:**

```python
import emoji
```

```python
comments['comment_text']
```

```python
all_emojis_list= [ ]
for comment in comments['comment_text']:
  for char in comment:
    if char in emoji.EMOJI_DATA:
      all_emojis_list.append(char)
```

```python
all_emojis_list[0:10]
```

```python
from collections import Counter
```

```python
Counter(all_emojis_list).most_common(10)
```

```python
emojis= [Counter(all_emojis_list).most_common(10)[i][0] for i in range(10)]
```

```python
emojis
```

```python
freqs= [Counter(all_emojis_list).most_common(10)[i][1] for i in range(10)]
```

```python
freqs
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
trace = go.Bar(x=emojis, y=freqs)
iplot([trace])
```
